```python

def predict(input_claim):
    api_key = '5b53715de78a4f0f9cce290a4144d444'
    api_endpoint = f"https://idir.uta.edu/claimbuster/api/v2/score/text/sentences/{input_claim}"
    request_headers = {"x-api-key": api_key}
    api_response = requests.get(url=api_endpoint, headers=request_headers)
    return api_response.json()['results'][0]['score']
```

```python
dataset = 'germeval'
path = f'/home/sami/Claimspotting/Datasets/ClaimbusterPredictions/{dataset}.csv'
df = pd.read_csv(path)
# list index out of range
for idx, row in df.iterrows():
    try: 
        float(row['prediction'])
    except ValueError as e:
        print(e)
        df.iloc[idx, 2] = predict(row['prediction'])
    
```

```python
def supplement_germeval(df):
    df_list = []
    models = ['logreg', 'svm']
    trainsets = ['checkthat2019', 'checkthat2021', 'checkthat2022', 'claimbuster']
    for model in models:
        for trainset in trainsets:
            row = [model, trainset, 'checkworthy', 'germeval'] + ([None]*(df.shape[1]-4))
            dct = {k:[v] for k,v in zip(df.columns, row)}
            frame_row = pd.DataFrame.from_dict(dct)
            df_list.append(frame_row)
    supplement = pd.concat(df_list, ignore_index=True)
    df = pd.concat([df, supplement], ignore_index=True)
    return df
```
